# Jeu de hazard
Créer une fonction de jeu de hazard.
* Le programme choisit aléatoirement un chiffre entre 1 et 10 inclus.
* Le joueur choisit aussi un chiffre entre 1 et 10 inclus.
* Si le joueur a gagné,
  * le message "Félicitations, vous êtes l'heureux gagnant du jeu concours !" est affiché
  * et la fonction renvoie la valeur de gain "500 €".
* Si le joueur a perdu,
  * le message "Dommage, ce sera pour une prochaine fois." est affiché
  * et la fonction renvoie la valeur de gain "0 €".
* Dans tous les cas, la fonction affichera également le message "Merci de votre participation.".
* A votre avis quelle est la probabilité que le joueur remporte les 500€ en jouant une seule fois ?

```python
# Jeu de hasard
# Import de random pour le tirage
import random

# Définition d'une fonction pour le tirage
def tirageHazard():
  return random.randint(1, 10)

# Définition d'une fonction pour le jeu de hasard
def jeuHazard():
  message = "erreur"
  gain = "erreur"
  choixUtilisateur = int(input("Veuillez entrez une valeur entre 1 et 10 :"))
  if choixUtilisateur <= 10 and choixUtilisateur >= 1:
    tirageProgramme = tirageHazard()
    if choixUtilisateur > tirageProgramme:
      message = "Félicitations, vous êtes l'heureux gagnant du jeu concours !"
      gain = "500 €"
    else:
      message = "Dommage, ce sera pour une prochaine fois."
      gain = "0 €"
  else:
    print("La valeur inscrite n'est pas contenue entre 1 et 10.")

  print(f"{message} Montant : {gain}. Merci de votre participation.")

jeuHazard()
```

# Compte à rebours

Créer un programme qui effectue un compte à rebours :
* En commençant par la valeur renseignée par l'utilisateur.
* En terminant à 0.
* Qui affiche le message "Il reste x opérations", x étant le nombre d'opérations restantes.
* Le message s'affiche uniquement lorsque x est un multiple de 5. Sinon aucun message n'est affiché.

```python
# Définition d'une fonction pour le compte à rebours
def compteRebours():
  valeurUtilisateur = int(input("Veuillez entrer une valeur numérique :"))
  i = valeurUtilisateur
  if i % 5 == 0:
    for i in range(valeurUtilisateur, 0, -1):
      i = i
      print(f"Il reste {i} opérations")
      i = i - 1
  else:
    print("Ah oui, ce doit être un multiple de 5 enfait...")

compteRebours()
```

#Distance 3D

* Créez une fonction calc_distance_3D() qui calcule la distance euclidienne en trois dimensions entre deux atomes.  
* Testez votre fonction sur les 2 points A(0,0,0) et B(1,1,1). Trouvez-vous bien √3 ?  
* On rappelle que la distance euclidienne d entre deux points A et B de coordonnées cartésiennes respectives (xA, yA,zA) et (xB, yB,zB) se calcule comme suit :  

formule_3D.png

```python
# Importation de la librairie math, et on va en avoir besoin...
import math
# Définition de la fonction de calcule de distance
def calc_distance_3d(xA, yA, zA, xB, yB, zB):
  # Traduction de la formule
  distanceEuclidienne = math.sqrt((xB - xA)**2 + (yB - yA)**2 + (zB - zA)**2)
  return distanceEuclidienne

# Les valeurs de test pour A et B :
xA, yA, zA = 0, 0, 0
xB, yB, zB = 1, 1, 1

# Test de la fonction
distance = calc_distance_3d(xA, yA, zA, xB, yB, zB)

# On croise les doigts
print(f"La distance vaut {distance}")
```

# Distance de Hamming

La distance de Hamming mesure la différence entre deux séquences de même taille en comptant le nombre de positions
qui, pour chaque séquence, ne correspondent pas au même acide aminé.
* Créer la fonction `dist_hamming()` qui prend en argument deux chaînes de caractères et qui renvoie la distance de
Hamming (sous la forme d’un entier) entre ces deux chaînes de caractères.
* Calculez la distance de Hamming entre les séquences
`AGWPSGGASAGLAIL` et `IGWPSAGASAGLWIL`
* puis entre les séquences
`ATTCATACGTTACGATT` et `ATACTTACGTAACCATT`.

Exemple : la distance de Hamming entre
* "ramer" et "cases" est égale à 3.

```python
# Définition d'une fonction pour mesurer la différence entre deux séquences
def distanceHamming(sequence1, sequence2):
  # Le compteur initialisé à 0
  distance = 0
  # Petite condition dans le cas où l'utilisateur (moi) me trompe dans l'entrée des séquences
  if len(sequence1) == len(sequence2):
    # La boucle pour compter
    for i in range(len(sequence1)):
      if sequence1[i] != sequence2[i]:
        distance = distance + 1
    return distance
  else:
    print(f"Votre séquence {sequence1} n'a pas la même longueur que {sequence2}")

# Teste pour les séquences AGWPSGGASAGLAIL et IGWPSAGASAGLWIL
seq1 = "AGWPSGGASAGLAIL"
seq2 = "IGWPSAGASAGLWIL"
distance1 = distanceHamming(seq1, seq2)
print(f"La distance du premier test est de {distance1}")
# Teste pour les séquences ATTCATACGTTACGATT et ATACTTACGTAACCATT
seq3 = "ATTCATACGTTACGATT"
seq4 = "ATACTTACGTAACCATT"
distance2 = distanceHamming(seq3, seq4)
print(f"La distance du second test est de {distance2}")

```

#Traducteur chiffre Romains

Ecrire une fonction qui prend un nombre entier compris entre 1 et 3000 inclus, et le convertit en chiffre romains.  
Ci-dessous la correspondance :
https://jeretiens.net/wp-content/uploads/2015/12/les_chiffres_romains.jpg
![Exemple de chiffres romains](https://jeretiens.net/wp-content/uploads/2015/12/les_chiffres_romains.jpg)



```python
# Définition d'une fonction de traduction en chiffre romains
def convertusCaius(valeurUtilisateur):
  resultat = ""
  valeurs = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
  symboles = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']
  i = 0
  if valeurUtilisateur < 1 or valeurUtilisateur > 3000:
    message = "Valeur non comprise entre 1 et 3000"
    return message
  else:
    # On va parcourir la liste de valeurs et ajouter autant de symboles possibles tant que la valeur donnée est supérieure aux valeurs de cette liste
    for i in range(len(valeurs)):
      while valeurUtilisateur >= valeurs[i]:
        resultat = resultat + symboles[i]
        valeurUtilisateur = valeurUtilisateur - valeurs[i]
    message = f"Ce qui donnera {resultat} en chiffre romains"
    return message

valeurUtilisateur = int(input("Veuillez entrer une valeur entre 1 et 3000"))
convertusCaius(valeurUtilisateur)
```
